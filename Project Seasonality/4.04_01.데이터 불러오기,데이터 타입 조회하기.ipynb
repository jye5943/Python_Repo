## p30 함수 만들고 이상 데이터 정제하기

```python
#컬럼 추가 생성 및 조건에 따른 컬럼명 추가
def refining(df):
    if df['PRODUCTAGE'] < 1:
        return 1
    elif df['PRODUCTAGE'] < 2:
        return 2
    else:
        return 5
```

```python
customerData["productage_new2"] = customerData.apply(refining, axis=1)
customerData.head()
```

```python
SortKey = ["CUSTID","AVERAGEPRICE"]
```

```python
# ascending -true:오름차순, false:내림차순(기본값true) 
# inplace- true: 변경사항 저장, false: 조회만o저장x(기본값false)
customerData.sort_values(SortKey, ascending=[True,False], inplace=True)
customerData.head()
```

## 이상 데이터 조회하기 (실습)

```python
import numpy as np
```

```python
# \표시는 윗줄이랑 아랫줄이 같은 줄이라는걸 나타냄
customerData["PRODUCTAGE_NEW"] = np.where(customerData["PRODUCTAGE"] < 1, 1,\
                                          np.where(customerData["PRODUCTAGE"]<2, 2,\
                                          np.where(customerData["PRODUCTAGE"]<3, 3,\
                                          5)))
customerData.head()
```

```python
# window 범위, center=ture일때 나를 중간 값으로 5개를 묶는다.
# center=false일때 내가 마지막이 되어 5개를 묶는다.)
# NaN : none 계산할 수 없다
#min_period : 계산할 수 있는 최소한의 범위
```

```python
customerData["TEST"] = customerData["PRODUCTAGE"].rolling(window = 5, min_periods=1, center = False).mean()
customerData.head()
```

```python
customerData["TEST"] = customerData["PRODUCTAGE"].rolling(window = 5, center = True).mean()
customerData
```
