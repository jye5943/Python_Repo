## 지도학습 - Regression

## 1. import & 데이터 불러오기

```python
import pandas as pd
import numpy as np
from sklearn import datasets, tree  #decision tree를 쓸거기 때문에 tree 라이브러리를 불러옴
# 지금 데이터 셋을 안쓰는 이유는 ? 데이터 셋 : 무작위로 70퍼를 뽑아서 예측을 함. 이 때 만약에 무작위인 경우에 미래를 뽑아서 예측을 하다 보면
# 과거를 이용해 최신 미래 데이터를 학습을 해야지 정확도 있는 데이터를 뽑아낼 수 있기 때문
```

```python
# 모델 정확도 라이브러리 선언
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
import math
```

```python
featuresData =pd.read_csv("../dataset/feature_regression_example.csv")
featuresData.head()
```

## 2. 데이터 정제하기 

```python
#홀리데이 및 프로모션 여부 컬럼에 대해서 Y-> 1 , N -> 0 컬럼을 추가로 생성하기
featuresData["HO_YN"] = np.where(featuresData["HOLIDAY"] == "Y", 1, 0)
featuresData["PRO_YN"] = np.where(featuresData["PROMOTION"] == "Y", 1, 0)
```

```python
featuresData.head()
```

```python
#데이터 범위를 YEARWWEK 기준 201501~201652 구간만 조회하여 FeaturesData 변수에 저장하시오.

#방법1 (속성 값 이용하기)
featuresData = featuresData[(featuresData.YEARWEEK >= 201501) & (featuresData.YEARWEEK <= 201652)]

#방법2 (인덱스 이용하기)
# FeaturesData = featuresData[(featuresData["YEARWEEK"] >= 201501) & (featuresData["YEARWEEK"] <= 201652)]
# FeaturesData.head()
```

```python
featuresData.head()
```

```python
# 상관관계 확인 (데이터 프레임명.corr() )
corrDf = featuresData.corr()
corrDf
```

```python
# 인덱스컬럼 초기화
corrResult = corrDf.reset_index()
corrResult
```

```python
featuresData.head()
```

### 상관관계 확인 (데이터프레임.corr())

```python
# 상관관계 계수 비율 선정
featuresStd = 0.6  # 60퍼센트
```

```python
# feature 선택 (60퍼센트 이상인 자료를 뽑는 것)
features = corrResult[ (corrResult.QTY != 1) & 
(abs(corrResult.QTY) > featuresStd)]["index"].tolist()  #abs 절대값 
```

```python
features
```

```python
#label값은 그냥 변하지 않기 때문에 선언했다고 보면 됨
label = ["QTY"]
label
```

## 3. 예측

### 데이터 분리

```python
#데이터 분리
yearweekStd = 201630
```

```python
# 내가 원하는 컬럼 feature과 label만 뽑고 싶은 경우 
# 실제 특성하고 답지를 집어 넣고 실험을 해줘야함. 따라서 처음 지도를 할 때 1,2 를 넣고 실행을 시켜주어야 함

# 1. 학습시킬 featrues
trainingData_features = featuresData[featuresData.YEARWEEK <= yearweekStd][features]  #마지막에 [featrues] 는 뽑고싶은 컬럼 이름 

# 2. 학습시킬 정답지
trainingData_label = featuresData[featuresData.YEARWEEK <= yearweekStd][label]

# 3. 테스트 할 features 
testData_features = featuresData[featuresData.YEARWEEK > yearweekStd][features]

# 4. 테스트할 정답지 
testData_label = featuresData[featuresData.YEARWEEK > yearweekStd]  # 여기에 merge시킬거여서 인덱스를 안 붙임
```

### 1) DecisionTree Regression으로 예측하기

```python
# 모델 정의
model_method = tree.DecisionTreeRegressor()
# 모델에 특성과 라벨 값을 넣어준다. (Feature & Label)  ***여기서 중요한 것은 특성과 답지(label)를 함께 넣어서 학습을 시켜야 한다. 
model = model_method.fit(trainingData_features,trainingData_label)
#.fit : 학습을 시키는 함수(traingData_feautures와 trainingData_label을 던져주면 학습모델이 만들어지게 된다.
#features를 종속변수라고 하기도 함.
```

```python
# 내가 검증할 testData_features만 넣어 예측값 결과 도출
predict = model.predict(testData_features)
predict
```

### 데이터 정리

```python
predictData = pd.DataFrame(predict)
predictData["PREDICT"] = predictData[0] 
del(predictData[0])
predictData.head()
```

```python
testData_label.head()
```

### 인덱스 없애기 & 데이터 merge 하기 (concat, outer join)

```python
#인덱스를 없애주기
indexTextData = testData_label.reset_index(drop=True)

# 방법 1.concat 이용해서 data merge
resultConcat = pd.concat([indexTextData,predictData],axis=1)
resultConcat.head()
```

```python
# 방법 2. outer join을 이용해서 data merge
result=pd.merge(indexTextData, predictData,left_index=True, right_index=True, how='outer')
result.head()
```

```python
# 방법 3. 새 컬럼을 생성해서 데이터를 밀어 넣기(concat 이용) -> 인덱스를 삭제해줄 필요 없음.
predictData = pd.DataFrame(predict)
predictData.columns = ["PREDICT"]
predictData1 = pd.concat([testData_label.reset_index(drop=True), predictData], axis=1)
```

## 4. 검증

### 결과 검증

```python
# 1번과 2번의 결과가 비슷하면 신뢰할 수 있는 데이터이다. 
### 1. MAE :mean absolute error   평균 절대 오차 
mean_absolute_error(resultConcat['QTY'], resultConcat['PREDICT'])                
```

```python
math.sqrt(mean_squared_error(resultConcat['QTY'], resultConcat['PREDICT']))   #math.sqrt는 루트를 의미함.
```

```python
### 2. RMSE (MSE) : root mean square error 평균 제곱근 오차
mean_squared_error(resultConcat['QTY'], resultConcat['PREDICT'])
```

```python
### 3. R2 (결정계수)
r2_score(resultConcat['QTY'], resultConcat['PREDICT']) 
```
