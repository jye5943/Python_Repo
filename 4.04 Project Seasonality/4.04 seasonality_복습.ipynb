### 컬럼 추가 생성 및 조건에 따른 컬럼명 변경 함수만들기

```python
## refining 자리: 함수 이름 / (cu): 데이터 이름 지정 
def refining(cu):
    if cu["PRODUCTAGE"] < 1:
        return 1 
    elif cu["PRODUCTAGE"] < 2:
        return 2
    else :
        return 5
```

```python
customerData = pd.read_csv("../dataset/customerdata.csv")
```

```python
customerData.dtypes
```

##### axis=1 y축(열)을 추가, axis=0 x축(행)을 추가

```python
customerData["productage_new2"]=customerData.apply(refining,axis=1)
customerData.head()
```

##### ascending =True 오름차순, False 내림차순 / inplace=True 저장, False 저장안함

```python
customerData.sort_values(["CUSTID","AVERAGEPRICE"],ascending=True,inplace=True)
customerData.head()
```
