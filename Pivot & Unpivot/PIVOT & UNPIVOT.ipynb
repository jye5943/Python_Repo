##### 내가 가지고 있는 데이터를 올릴 때 컬럼을 안 작성하고 올리는 경우가 있다. (pivot식으로 자료를 받지 못하는 경우가 있다.) 이 때, unpivot을 시켜서 (melt)시켜서 컴퓨터와 대화할 수 있게 해준다. 
##### //pivot - 설명력 UP
##### //unpivot - 컴퓨터로 제어할 때 편한 형식

### PIVOT(행을 열로)

```python
import pandas as pd
```

```python
selloutData = pd.read_csv("./dataset/kopo_channel_seasonality_new.csv")
```

```python
pivotResult = selloutData.\
pivot_table(index = ["REGIONID","PRODUCT"], columns ='YEARWEEK' , values ='QTY', aggfunc ="mean")
pivotResult2 =pivotResult.reset_index()
```

```python
pivotResult2.head()
```

### UNPIVOT(열을 행으로)

```python
unpivotResult = pd.melt(
    pivotResult2, id_vars=['REGIONID', 'PRODUCT'], 
    var_name='YEARWEEK', value_name='QTY')
unpivotResult.head()
```
