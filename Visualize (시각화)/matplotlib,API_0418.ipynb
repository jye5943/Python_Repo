차트 꾸미기 (차트 선색 및 크기 설정)

```python
x=list(range(1,11,2))   #1~11사이에서 2의 간격을 준다

y=[]
for i in x: 
    y.append(i**2) #-> y축으로 i제곱을 하겠다는 의미 , object를 append는 맨 뒤에 추가함 
    
    ## ( = 위 3줄이랑 같은거 : y=[i**2 for i in x] )

plt.plot(x,y,'r',lw=5.4) #lw: line width, 선 굵기
```

```python
#필요 라이브러리 정의
import matplotlib.pyplot as plt

# 팝업 창 활용하여 차트 시연
x=list(range(1,11,1))
y = [i**2 for i in x]
y2 = [i**2.2 for i in x]

# 서브플롯
fig = plt.figure(figsize=(10,5))  #가로 10
ax1 = fig.add_subplot(211) #2행 1열의 1번째 (그래프의 위치)
plt.plot(x,y, 'b',lw=1.5, label = '1st')
plt.plot(x,y, 'o')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('subplot')
ax2 = plt.subplot(212) #2행1열의 2번째
plt.plot(x,y2, 'g', lw=1.5, label='2nd')
plt.plot(x,y2, 'go')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

```python
x=list(range(1,11,1))
y = [i**2 for i in x]
y2 = [i**2.2 for i in x]

# 이중축
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,1,1)
plt.plot(x,y, 'b--',lw=1.5, label = '1st')
plt.plot(x,y, 'o')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('subplot')
ax2 = ax1.twinx()
plt.plot(x,y2, 'g', lw=1.5, label='2nd')
plt.plot(x,y2, 'go')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

featuresData를 활용하여 
2015년도 자료만 활용하여
x축: week 이중축으로 qty, pro_percent를 
차트로 그리세요.

```python
featuresData = pd.read_csv("./dataset/feature_regression_example.csv")
featuresData2 = featuresData[(featuresData.YEARWEEK >=201501) & (featuresData.YEARWEEK <= 201552)]
x=featuresData2["WEEK"]
y=featuresData2["QTY"]
y2=featuresData2["PRO_PERCENT"]

fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,1,1) #1행1열 첫번째
plt.plot(x,y, 'bo--',lw=1.5, label = '1st')
plt.grid(True) ##격자무늬가 뒤에 나오도록 프린트
plt.legend(loc=0) 
plt.xlabel('WEEK')
plt.ylabel('QTY')  #x,ylabel은 plot선 한번 그을때 한번씩만 적을 수 있기 때문에 라벨이 여러개라면 선을 생성한 뒤 새로운 라벨명을 써줘야함.
plt.title('subplot')
ax2 = ax1.twinx() #1행1열 첫번째로-> 같은 테이블에 나오겠다
plt.plot(x,y2, 'g-', lw=1.5, label='2nd')
plt.legend(loc=0)
plt.ylabel('PRO_PERCENT')
```
